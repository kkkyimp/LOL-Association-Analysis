```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('LOL数据.csv',) 
```

```python
'''仅做队内数据'''
data1=data.iloc[:,0:5]
new_col = ['上', '野', '中', '下' , '辅']
data1.columns = new_col
data2=data.iloc[:,5:10]
new_col = ['上', '野', '中', '下' , '辅']
data2.columns = new_col
data=pd.concat([data1,data2],ignore_index=True)
```

```python
data
```

```python

data_1 = pd.get_dummies(data,columns=['上', '野', '中', '下' , '辅'])
```

```python
frequent_itemsets = apriori(data_1, min_support=0.001, use_colnames=True,max_len=2)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=3)
```

```python
rules
```

```python
pd.set_option('max_row',500) 
rules.sort_values(by='lift',ascending=False)
```

```python
out=rules[['antecedents','consequents','lift']]
```

```python
out1=out
```

```python
out11=out1.loc[:,['antecedents','consequents']].applymap(list)['antecedents']
out12=out1.loc[:,['antecedents','consequents']].applymap(list)['consequents']
```

```python

t1=pd.DataFrame([u.split('_') for u in   [str(i[0]) for i in out11]], columns=['wei1','name1']).applymap(str)
t2=pd.DataFrame([u.split('_') for u in   [str(i[0]) for i in out12]], columns=['wei2','name2']).applymap(str)
```

```python
data0=t1.join(t2).join(out['lift'])
```

```python
data0.to_csv('R数据.csv',sep=',',index=False,header=True)
```
